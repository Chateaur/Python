# Routes -- Seine Maritime

## Partie 1 - Chargement et nettoyage

Import des librairies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Chargement du csv

```python
df = pd.read_table('data/routesTravaux.csv', index_col = 0, sep = ";", decimal = ",")
df = df.reset_index()
df.head()
```

```python
df.dtypes
```

Le type epaisseurdernierstravaux est bien float64

```python
df.isnull().sum()
```

La colonne supportdernierstravaux contient 6 NaN. On les remplace :

```python
df = df.fillna(method='ffill')
```

```python
df.describe(include = 'all')
```

Remarques :
   * La colonne age est toujours nulle.
   * Des valeurs 9999,99 sont répétées dans le data set pour la colonne epaisseurdernierstravaux et sont très éloignées des autres valeurs (autour de 4). Ces valeurs peuvent témoigner d'une absence de valeur (à voir avec l'expert métier)

```python
df.loc[df['epaisseurdernierstravaux'] == 9999.99]
```

On voit que très souvent, une valeur de 9999.99 pour epaisseurdernierstravaux est associée avec des valeurs anormales ('!!!!!!') pour les colonnes naturedernierstravaux et.ou supportdernierstravaux. On supprime donc ces lignes. 

```python
df = df[df.epaisseurdernierstravaux != 9999.99]
df = df[df.naturedernierstravaux != '!!!!!!']
df = df[df.supportdernierstravaux != '!!!!!!']
```

## Partie 2 Analyse

### Fréquence de nature de travaux

```python
grouped = df['naturedernierstravaux'].groupby(df['naturedernierstravaux'])

counted = grouped.count()

counted.sort_values(ascending = False, inplace = True) #Ordonne la série

top5 = counted[:6]
print(top5)
```

Diagramme en barre de la répartition de la nature des travaux

```python
plt.xlabel('Nature travaux')
plt.ylabel('Nombre de travaux')
plt.title('Répartition de la nature des travaux')

plt.bar(range(len(counted)), counted, align='center')
plt.xticks(range(len(counted)), counted.index.str.lower(), rotation=80)

plt.grid(True)
plt.show()
```

### Analyse par année

```python
groupedByYear = pd.DataFrame({'Nombre de travaux' : df.groupby('anneedernierstravaux').size(),
                              'Epaisseur derniers travaux moyenne' : df.groupby('anneedernierstravaux')['epaisseurdernierstravaux'].mean().round(2)})
print(groupedByYear)
```

Courbes :

```python
plt.figure(figsize=(10,5))

plt.subplot(121)
plt.title('Epaisseur moyenne\n derniers travaux')
plt.plot(groupedByYear.index, groupedByYear['Epaisseur derniers travaux moyenne'], 'b.-')

plt.subplot(122)
plt.title('Nombre de travaux\n par année')
plt.plot(groupedByYear.index, groupedByYear['Nombre de travaux'], 'r.-')

plt.show()
```

On remarque une tendance de l'épaisseur moyenne des derniers travaux à diminuer alors que le nombre de travaux augmente. Cela peut se traduire par le fait que le département de Seine Maritime a tendance à réaliser des chantiers plus petits mais plus nombreux qu'auparavant.
